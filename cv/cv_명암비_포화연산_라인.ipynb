```python
import numpy as np
import cv2

def drawLines():
    img = np.full((400, 400, 3), 255, np.uint8)

    cv2.line(img, (50, 50), (200, 50), (0, 0, 255))
    cv2.line(img, (50, 100), (200, 100), (255, 0, 255), 3)
    cv2.line(img, (50, 150), (200, 150), (255, 0, 0), 10)

    cv2.line(img, (250, 50), (350, 100), (0, 0, 255), 1, cv2.LINE_4)
    cv2.line(img, (250, 70), (350, 120), (255, 0, 255), 1, cv2.LINE_8)
    cv2.line(img, (250, 90), (350, 140), (255, 0, 0), 1, cv2.LINE_AA)

    cv2.arrowedLine(img, (50, 200), (150, 200), (0, 0, 255), 1)
    cv2.arrowedLine(img, (50, 250), (350, 250), (255, 0, 255), 1)
    cv2.arrowedLine(img, (50, 300), (350, 300), (255, 0, 0), 1, cv2.LINE_8, 0, 0.05)

    cv2.drawMarker(img, (50, 350), (0, 0, 255), cv2.MARKER_CROSS)
    cv2.drawMarker(img, (100, 350), (0, 0, 255), cv2.MARKER_TILTED_CROSS)
    cv2.drawMarker(img, (150, 350), (0, 0, 255), cv2.MARKER_STAR)
    cv2.drawMarker(img, (200, 350), (0, 0, 255), cv2.MARKER_DIAMOND)
    cv2.drawMarker(img, (250, 350), (0, 0, 255), cv2.MARKER_SQUARE)
    cv2.drawMarker(img, (300, 350), (0, 0, 255), cv2.MARKER_TRIANGLE_UP)
    cv2.drawMarker(img, (350, 350), (0, 0, 255), cv2.MARKER_TRIANGLE_DOWN)

    cv2.imshow("img", img)
    cv2.waitKey()
    cv2.destroyAllWindows()
    
if __name__ == "__main__" :
    drawLines()
```

```python
# 이벤트로 박스 그리기
import cv2 as cv
import sys
file_path='./img/girl_laughing.jpg'
img=cv.imread(file_path)
  
if img is None:
    sys.exit('파일을 찾을 수 없습니다.')
    
def draw(event,x,y,flags,param): # 정해진 양식 => 내가 활용하진 않더라도 setMouseCallBack이 요구함
    global ix,iy
    
    if event==cv.EVENT_LBUTTONDOWN:	# 마우스 왼쪽 버튼 클릭했을 때 초기 위치 저장
        ix,iy=x,y
        print(ix, iy)
    elif event==cv.EVENT_LBUTTONUP:	# 마우스 왼쪽 버튼 클릭했을 때 직사각형 그리기
        # x,y를 반환하나? => OpenCV에서 자동으로 전달되는 인자, 마우스 버튼이 떼어진 시점의 커서 좌표 (x, y)를 반환
        print(x,y)
        cv.rectangle(img,(ix,iy),(x,y),(0,0,255),2)
    
    cv.imshow('Drawing',img)
    
#함수 밖의 cv.imshow()는 초기 이미지 창을 띄우기 위해 필요
#함수 안의 cv.imshow()는 이미지가 업데이트될 때마다 다시 그려주기 위해 필요
    
cv.namedWindow('Drawing') # Drawing이라는 창에서
cv.imshow('Drawing',img) # Drawing이라는 창에서 img 올리고

cv.setMouseCallback('Drawing',draw) # Drawing이라는 창에서 draw라는 함수를 작동하게 해라

while(True):
    if cv.waitKey(1)==ord('q'):
        cv.destroyAllWindows()      
        break
```

```python
import cv2 as cv
from PIL import ImageDraw, ImageFont, Image
import numpy as np

file_path = "./img/blank_500.jpg"
img=cv.imread(file_path)
image_pil = Image.fromarray(cv.cvtColor(img, cv.COLOR_BGR2RGB))
font = ImageFont.truetype('NotoSansKR-ExtraBold.ttf', 40)
draw = ImageDraw.Draw(image_pil)
text = "심재성"
draw.text((img.shape[0]//2-60, img.shape[1]//2 - 60), text, font=font, fill=(255, 0, 0))
img = cv.cvtColor(np.array(image_pil), cv.COLOR_RGB2BGR)

def draw_circle(event,x,y,flags,param):
    global ix,iy
    if event == cv.EVENT_LBUTTONUP :
        cv.circle(img,(x,y),30,(0,0,255),-1)
        print(event)
    elif event == cv.EVENT_RBUTTONDOWN :
        ix,iy = x,y
        print(event)
    elif event == cv.EVENT_RBUTTONUP :
        cv.rectangle(img,(ix,iy),(x,y),(0,0,0),-1)
        print(event)
    cv.imshow("circle_example",img)

cv.namedWindow("circle_example",cv.WINDOW_AUTOSIZE)
cv.imshow("circle_example",img)

cv.setMouseCallback("circle_example",draw_circle)

while True :
    if cv.waitKey(1) == ord('q') :
        cv.destroyAllWindows()
        break
```

```python
import cv2

title = 'mouse event'                   # 창 제목
img = cv2.imread('./img/blank_500.jpg') # 백색 이미지 읽기
cv2.imshow(title, img)                  # 백색 이미지 표시

colors = {'black':(0,0,0),
         'red' : (0,0,255),
         'blue':(255,0,0),
         'green': (0,255,0) } # 색상 미리 정의

def onMouse(event, x, y, flags, param): # 콜백 함수 구현 ---①
    print(event, x, y, flags)                # 파라미터 출력
    color = colors['black']
    if event == cv2.EVENT_LBUTTONDOWN:  # 왼쪽 버튼 누름인 경우 ---②
        # 컨트롤키와 쉬프트 키를 모두 누른 경우 
        if flags & cv2.EVENT_FLAG_CTRLKEY and flags & cv2.EVENT_FLAG_SHIFTKEY : # flags를 같이 전달해야하며 & 비트연산자로 작동함. 논리연산자 작동X
            color = colors['green']
        elif flags & cv2.EVENT_FLAG_SHIFTKEY : # 쉬프트 키를 누른 경우
            color = colors['blue']
        elif flags & cv2.EVENT_FLAG_CTRLKEY : # 컨트롤 키를 누른 경우
            color = colors['red']
        # 지름 30 크기의 검은색 원을 해당 좌표에 그림
        cv2.circle(img, (x,y), 30, color, -1) 
        cv2.imshow(title, img)          # 그려진 이미지를 다시 표시 ---③

cv2.setMouseCallback(title, onMouse)    # 마우스 콜백 함수를 GUI 윈도우에 등록 ---④

while True:
    if cv2.waitKey(1) == ord('q'):     # esc로 종료
        break
cv2.destroyAllWindows()
```

```python
# 트랙바

import cv2
import numpy as np

win_name = 'Trackbar'                                   # 창 이름

img = cv2.imread('./img/blank_500.jpg')
cv2.namedWindow(win_name)
# cv2.imshow(win_name,img)                                # 초기 이미지를 창에 표시

# 트랙바 이벤트 처리 함수 선언 ---①
def onChange(x):                                        
    print(x)                                            # 트랙바 새로운 위치 값 --- ②
    # 'R', 'G', 'B' 각 트랙바 위치 값    --- ③
    r = cv2.getTrackbarPos('R',win_name)               
    g = cv2.getTrackbarPos('G',win_name)               
    b = cv2.getTrackbarPos('B',win_name)               
    print(r, g, b)
    # print(img)
    img[:] = [b,g,r]                                    # 기존 이미지에 새로운 픽셀 값 적용 --- ④
    # print(img)
    cv2.imshow(win_name, img)                           # 새 이미지 창에 표시

# 트랙바 생성    --- ⑤

cv2.createTrackbar('R', win_name, 255, 255, onChange)  
cv2.createTrackbar('G', win_name, 255, 255, onChange)
cv2.createTrackbar('B', win_name, 255, 255, onChange)
cv2.imshow(win_name,img)
while True:
    if cv2.waitKey(1) == ord('q'):
        break
cv2.destroyAllWindows()   
```
