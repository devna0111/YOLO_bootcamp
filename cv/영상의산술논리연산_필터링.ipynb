```python
import sys
import numpy as np
import cv2
from matplotlib import pyplot as plt


src1 = cv2.imread('img/lenna256.bmp', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('img/square.bmp', cv2.IMREAD_GRAYSCALE)

if src1 is None or src2 is None:
    print('Image load failed!')
    sys.exit()

dst1 = cv2.bitwise_and(src1, src2)
dst2 = cv2.bitwise_or(src1, src2)
dst3 = cv2.bitwise_xor(src1, src2)
dst4 = cv2.bitwise_not(src1)

plt.subplot(231), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('src1')
plt.subplot(232), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('src2')
plt.subplot(233), plt.axis('off'), plt.imshow(dst1, 'gray'), plt.title('bitwise_and')
plt.subplot(234), plt.axis('off'), plt.imshow(dst2, 'gray'), plt.title('bitwise_or')
plt.subplot(235), plt.axis('off'), plt.imshow(dst3, 'gray'), plt.title('bitwise_xor')
plt.subplot(236), plt.axis('off'), plt.imshow(dst4, 'gray'), plt.title('bitwise_not')
plt.show()
```

```python
# 200 * 400 size  and, or, xor, not 
from PIL import Image
data1 = np.full((100,400),0)
data2 = np.full((100,400),255)
img2 = np.vstack((data1,data2)).astype(np.uint8)

data1 = np.full((200,200),255)
data2 = np.full((200,200),0)
img1 = np.hstack((data1,data2)).astype(np.uint8)

fig, axes = plt.subplots(3, 2, figsize=(9, 8))

axes[0,0].imshow(img1, 'gray')
axes[0,1].imshow(img2, 'gray')

and_ = cv2.bitwise_and(img2,img1)
axes[1,0].imshow(and_, 'gray')

or_ = cv2.bitwise_or(img2,img1)
axes[1,1].imshow(or_, 'gray')

xor_ = cv2.bitwise_xor(img2,img1)
axes[2,0].imshow(xor_, 'gray')

not_img1 = img1[:,::-1]
axes[2,1].imshow(not_img1, 'gray')

axes[0, 0].set_title("img1")
axes[0, 1].set_title("img2")
axes[1, 0].set_title("and")
axes[1, 1].set_title("or")
axes[2, 0].set_title("xor")
axes[2, 1].set_title("not(img1)")

plt.show()
```

```python
import numpy as np
import cv2 

robot1 = cv2.imread('img/robot_arm1.jpg')
robot2 = cv2.imread('img/robot_arm2.jpg')

dif = np.where(robot1!=robot2,255,0).astype(np.uint8)
dif[:,:,:-1] = 0 # xor  

dif_robo = cv2.bitwise_or(dif, robot1)

cv2.imshow('dif',dif)
cv2.imshow('dif_robo',dif_robo)
cv2.waitKey()
cv2.destroyAllWindows()
```
