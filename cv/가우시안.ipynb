```python
x = np.linspace(-5,5,100)
x_std = x.std() ; x_var = x.var()
# x_std, x_var, np.pi, np.exp(1)
y = (np.exp(1.0)**((-1*(x-x.mean())**2)/(2*x_var))) / (((2*np.pi)**0.5)*x_std)
# std = 2.0 // std = 4.0 버젼 추가
x_std2 = 2.0 ; x_std3 = 4.0
y2 = (np.exp(1.0)**((-1*(x-x.mean())**2)/(2*x_std2**2))) / (((2*np.pi)**0.5)*x_std2)
y3 = (np.exp(1.0)**((-1*(x-x.mean())**2)/(2*x_std3**2))) / (((2*np.pi)**0.5)*x_std3)
plt.figure(figsize=(4,3))

plt.plot(x,y, color = 'k')
plt.plot(x,y2, color = 'b')
plt.plot(x,y3, color = 'g')

plt.show()
```
