```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
x = np.linspace(-5,5,10)
y = 3*x + 1
plt.figure(figsize=(6,4))
plt.plot(y, label="y값", marker='o')
plt.ylim(0,10)
plt.xlim(0,10)
plt.legend(loc=(1.05,0.9))
plt.title("y=3x+1 그래프")
plt.show()
```

```python
def f(x) :
    return x**3 + 1

def g(x) :
    return (x+2)**0.5

print("(g o f)(2)=",g(f(2)))
print("(f o g)(2)=",f(g(2)))
print("(g o g)(2)=",g(g(2)))
```
